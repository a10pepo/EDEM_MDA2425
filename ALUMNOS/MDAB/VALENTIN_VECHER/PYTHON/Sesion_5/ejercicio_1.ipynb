<span style='color:lightblue'>CREAR UNA COLUMNA EXTRA CALCULADA</span>

<span style='color:lightblue'>La columna debe ser la suma de HP + ATAQUE + DEFENSA + VELOCIDAD</span>

<span style='color:lightblue'>La columna debe llamarse TOTAL</span>

```python
pk['Total']= pk[['HP','Attack','Defense','Speed']].sum(axis=1)
print(pk.head(5))
```

<span style='color:lightblue'>ELIMINAR LA COLUMNA TOTAL</span>


```python
pk.drop('Total', axis=1)
```

<span style='color:lightblue'>FILTRAR POKEMONS DE TIPOS "GRASS" Y "POISON"</span>


estos son los pokemons que tienen de tipo 1: Grass y de tipo 2: Poison, mediante (and)

```python
for indice, row in pk.iterrows():
    if row['Type 1']=='Grass' and row['Type 2']=='Poison':
        print(f'El pokemon {row['Name']} con indice {indice} es de tipo {row['Type 1']} de tipo 1 y de tipo {row["Type 2"]} de tipo 2')
    
```

<span style='color:lightblue'>FILTRAR POKEMONS DE TIPO "FIRE" Ó "POISON"</span>


```python
for indice, row in pk.iterrows():
    if row['Type 1']=='Fire' or row['Type 1']=='Poison':
        print(f'{row["Name"]} es de tipo {row["Type 1"]}')
        
```

<span style='color:lightblue'>FILTRAR POKEMONS DE TIPO "GRASS" Y "POISON" Y UN HP >= 70</span>


```python
for indice, row in pk.iterrows():
    if row['Type 1']=='Grass' and row['Type 2']=='Poison' and row['HP']>=70:
        print(f'{row['Name']} es de tipo 1 {row['Type 1']}, de tipo 2 {row["Type 2"]} y tiene 70 puntos de salud o más')
```

```python
pk_filtrado=pk[(pk['Type 1']=='Grass') & (pk['Type 2']=='Poison') & (pk['HP']>70)]

columns_show=['Name', 'Type 1', 'Type 2', 'HP']
print(pk_filtrado[columns_show])
```

<span style='color:lightblue'>FILTRAR POKEMONS CON NOMBRE "MEGA"</span>


```python
pk_fil_name=pk[(pk['Name'].str.contains('MEGA', case=False))]
print(pk_fil_name)
```

<span style='color:lightblue'>FILTRAR POKEMONS SIN NOMBRE "MEGA"</span>


```python
pk_fil_nomega=pk[~(pk['Name'].str.contains('MEGA', case=False, ))]
print(pk_fil_nomega)
```

<span style='color:lightblue'>FILTRAR POKEMONS CUYOS NOMBRES COMIENCEN CON "PI"</span>


```python
pk_fil_pi=pk[(pk['Name'].str.startswith('Pi'))]
print(pk_fil_pi)
```

<span style='color:lightblue'>RENOMBRADO DE COLUMNA "FIRE" a "FLAME"</span>


```python
pk.replace('Fire', 'Flame', inplace=True)
print(pk)
```

<span style='color:lightblue'>RENOMBRAR DE NUEVO A "FIRE" LA COLUMNA "FLAME" RECIÉN CAMBIADA</span>


```python
pk.replace('Flame', 'Fire', inplace=True)
pk.head(10)
```

<span style='color:lightblue'>CAMBIAR A TODOS LOS POKEMON LEGENDARIOS A TIPO "FIRE"</span>


```python
pk.loc[pk['Legendary']==True, 'Type 1']='Fire'
pk.loc[pk['Legendary']==True, 'Type 2']='Fire'
print(pk)
```

<span style='color:lightblue'>(Agrupación - groupBy) ESTADÍSTICAS DE MEDIA POR TIPO DE POKEMON y ORDENADOS POR DEFENSA</span>


```python
pk_group=pk.groupby('Type 1').mean(numeric_only=True).sort_values(by='Defense', ascending=False)

print(pk_group)
```

<span style='color:lightblue'>(Agrupación - groupBy) ESTADÍSTICAS DE MEDIA POR TIPO DE POKEMON y ORDENADOS POR ATAQUE</span>


```python
pk_group_attk=pk.groupby('Type 1').mean(numeric_only=True).sort_values(by='Attack', ascending=False)
print(pk_group_attk)
```

<span style='color:lightblue'>(Agrupación - groupBy) ESTADÍSTICAS DE MEDIA POR TIPO DE POKEMON y ORDENADOS POR HP</span>


```python
pk_group_attk=pk.groupby('Type 1').mean(numeric_only=True).sort_values(by='HP', ascending=False)
print(pk_group_attk)
```

<span style='color:lightblue'>(Agrupación - groupBy) ESTADÍSTICAS DE SUMA POR TIPO DE POKEMON</span>


```python
pk_group_sum=pk.groupby('Type 1').sum(numeric_only=True)
print(pk_group_sum)
```

<span style='color:lightblue'>(Agrupación - groupBy) ESTADÍSTICAS DE NÚMERO DE POKEMONS POR TIPO 1 DE POKEMON</span>


```python
conteo_por_tipo = pk.groupby('Type 1').size()
conteo_por_tipo_df = conteo_por_tipo.reset_index(name='Cantidad')

print(conteo_por_tipo_df)
```

<span style='color:lightblue'>(Agrupación - groupBy) ESTADÍSTICAS DE NÚMERO DE POKEMONS POR TIPO 1 y 2 DE POKEMON</span>


```python
count_by_type = pk.groupby(['Type 1', 'Type 2']).size().reset_index(name='Count')

print(count_by_type)


```
