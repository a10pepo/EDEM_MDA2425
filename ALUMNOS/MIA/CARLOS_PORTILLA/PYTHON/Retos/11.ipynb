```python
df_centrifuga['DateTime'] = pd.to_datetime(df_centrifuga['DateTime'], errors='coerce')
df_horas_centrifugas['DATEVALUE'] = pd.to_datetime(df_horas_centrifugas['DATEVALUE'], errors='coerce')

# Filtrar las operaciones 1 y 2
df_operacion_1_2 = df_horas_centrifugas[df_horas_centrifugas['Operación'].isin([1, 2])]

# Crear un DataFrame vacío para almacenar los resultados filtrados
df_resultado = pd.DataFrame()

# Recorrer cada fecha de operación 1 y 2
for i in range(0, len(df_operacion_1_2), 2):
    # Definir las fechas de inicio y fin de cada operación
    inicio = df_operacion_1_2.iloc[i]['DATEVALUE']
    fin = df_operacion_1_2.iloc[i + 1]['DATEVALUE'] if i + 1 < len(df_operacion_1_2) else inicio
    
    # Filtrar los datos de la centrífuga que estén dentro de ese intervalo de tiempo
    df_filtrado = df_centrifuga[(df_centrifuga['DateTime'] >= inicio) & (df_centrifuga['DateTime'] <= fin)]
    
    # Agregar los datos filtrados al DataFrame resultado
    df_resultado = pd.concat([df_resultado, df_filtrado])

# Mostrar el resultado final con todos los intervalos filtrados
print(df_resultado)
```
