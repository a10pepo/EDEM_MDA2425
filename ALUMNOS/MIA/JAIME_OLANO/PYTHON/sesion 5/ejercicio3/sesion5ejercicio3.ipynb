Ejercicio 3
Crea clases de automóvil distintas como y que dispongan de distintos atributos, pero hereden los métodos de Automóvil a la hora de Arrancar, Acelerar, Frenar o Parar, salvo que algunos deben tener más potencia que otros:
Coche
Moto
Camión

```python
class Automovil:
    def __init__(self, marca, modelo, velocidad_maxima):
        self.marca = marca
        self.modelo = modelo
        self.velocidad_maxima = velocidad_maxima
        self.velocidad_actual = 0
        self.arrancado = False
    
    def arrancar(self):
        if not self.arrancado:
            self.arrancado = True
            print(f"{self.marca} {self.modelo} ha arrancado.")
        else:
            print(f"{self.marca} {self.modelo} ya está en marcha.")
    
    def acelerar(self):
        if self.arrancado:
            self.velocidad_actual += 10
            if self.velocidad_actual > self.velocidad_maxima:
                self.velocidad_actual = self.velocidad_maxima
            print(f"{self.marca} {self.modelo} aceleró a {self.velocidad_actual} km/h.")
        else:
            print(f"{self.marca} {self.modelo} no puede acelerar porque está apagado.")
    
    def frenar(self):
        if self.velocidad_actual > 0:
            self.velocidad_actual -= 10
            print(f"{self.marca} {self.modelo} frenó a {self.velocidad_actual} km/h.")
        else:
            print(f"{self.marca} {self.modelo} ya está detenido.")
    
    def parar(self):
        if self.arrancado:
            self.arrancado = False
            self.velocidad_actual = 0
            print(f"{self.marca} {self.modelo} se ha detenido por completo.")
        else:
            print(f"{self.marca} {self.modelo} ya está apagado.")

# Clases derivadas con diferentes potencias
template = "{marca} {modelo} aceleró a {velocidad} km/h."
class Coche(Automovil):
    def acelerar(self):
        if self.arrancado:
            self.velocidad_actual += 15  # Más rápido que el estándar
            if self.velocidad_actual > self.velocidad_maxima:
                self.velocidad_actual = self.velocidad_maxima
            print(template.format(marca=self.marca, modelo=self.modelo, velocidad=self.velocidad_actual))
        else:
            print(f"{self.marca} {self.modelo} no puede acelerar porque está apagado.")

class Moto(Automovil):
    def acelerar(self):
        if self.arrancado:
            self.velocidad_actual += 20  # Moto más ágil
            if self.velocidad_actual > self.velocidad_maxima:
                self.velocidad_actual = self.velocidad_maxima
            print(template.format(marca=self.marca, modelo=self.modelo, velocidad=self.velocidad_actual))
        else:
            print(f"{self.marca} {self.modelo} no puede acelerar porque está apagado.")

class Camion(Automovil):
    def acelerar(self):
        if self.arrancado:
            self.velocidad_actual += 5  # Más lento al acelerar
            if self.velocidad_actual > self.velocidad_maxima:
                self.velocidad_actual = self.velocidad_maxima
            print(template.format(marca=self.marca, modelo=self.modelo, velocidad=self.velocidad_actual))
        else:
            print(f"{self.marca} {self.modelo} no puede acelerar porque está apagado.")

# Prueba de los vehículos
coche = Coche("Toyota", "Corolla", 180)
moto = Moto("Yamaha", "R1", 300)
camion = Camion("Volvo", "FH16", 120)

coche.arrancar()
coche.acelerar()
coche.frenar()
coche.parar()

moto.arrancar()
moto.acelerar()
moto.acelerar()
moto.frenar()
moto.parar()

camion.arrancar()
camion.acelerar()
camion.acelerar()
camion.frenar()
camion.parar()
```
