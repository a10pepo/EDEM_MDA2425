POKEMONS DE TIPO 1 == WATER

```python
water_pokemon = df[df["Type 1"] == "Water"]
print(water_pokemon)
```

ORDENACIÓN POR NOMBRE ASCENDENTE

```python
df_sorted = df.sort_values(by="Name", ascending=True)
print(df_sorted)
```

CREAR UNA COLUMNA EXTRA CALCULADA

La columna debe ser la suma de HP + ATAQUE + DEFENSA + VELOCIDAD

La columna debe llamarse TOTAL

```python
df["TOTAL"] = df["HP"] + df["Attack"] + df["Defense"] + df["Speed"]
print(df)
```

ELIMINAR LA COLUMNA TOTAL

```python
df = df.drop(columns=["TOTAL"])
print(df)
```

FILTRAR POKEMONS DE TIPOS "GRASS" Y "POISON"

```python
grass_poison_pokemon = df[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison")]
print(grass_poison_pokemon)
```

FILTRAR POKEMONS DE TIPO "FIRE" Ó "POISON

```python
fire_or_poison_pokemon = df[(df["Type 1"] == "Fire") | (df["Type 2"] == "Poison")]
print(fire_or_poison_pokemon)
```

FILTRAR POKEMONS DE TIPO "GRASS" Y "POISON" Y UN HP >= 70

```python
grass_poison_hp_pokemon = df[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] >= 70)]
print(grass_poison_hp_pokemon)
```

FILTRAR POKEMONS CON NOMBRE "MEGA"

```python
mega_pokemon = df[df["Name"].str.contains('Mega', case=False)]
print(mega_pokemon)
```

FILTRAR POKEMONS SIN NOMBRE "MEGA"

```python
mega_pokemon = df[~df["Name"].str.contains("Mega", case=False)]
print(mega_pokemon)
```

FILTRAR POKEMONS CUYOS NOMBRES COMIENCEN CON "PI"

```python
pi_pokemon = df[df["Name"].str.contains("^Pi", case=False)]
print(pi_pokemon)
```

RENOMBRADO DE COLUMNA "FIRE" a "FLAME"

```python
df["Type 1"] = df["Type 1"].replace("Fire", "Flame")
print(df)
```

RENOMBRAR DE NUEVO A "FIRE" LA COLUMNA "FLAME" RECIÉN CAMBIADA

```python
df["Type 1"] = df["Type 1"].replace("Flame", "Fire")
print(df)
```

CAMBIAR A TODOS LOS POKEMON LEGENDARIOS A TIPO "FIRE"

```python
df.loc[df["Legendary"] == True, "Type 1"] = "Fire"
print(df)
```

(Agrupación - groupBy) ESTADÍSTICAS DE MEDIA POR TIPO DE POKEMON y ORDENADOS POR DEFENSA

```python
df_numeric = df.select_dtypes(include="number")

media_tipo = df_numeric.groupby(df["Type 1"]).mean()

media_tipo_sorted_defensa = media_tipo.sort_values("Defense", ascending=False)

print(media_tipo_sorted_defensa)
```

(Agrupación - groupBy) ESTADÍSTICAS DE MEDIA POR TIPO DE POKEMON y ORDENADOS POR ATAQUE

```python
media_tipo_sorted_ataque = media_tipo.sort_values("Attack", ascending=False)

print(media_tipo_sorted_ataque)
```

(Agrupación - groupBy) ESTADÍSTICAS DE MEDIA POR TIPO DE POKEMON y ORDENADOS POR HP

```python
media_tipo_sorted_hp = media_tipo.sort_values("HP", ascending=False)

print(media_tipo_sorted_hp)
```

(Agrupación - groupBy) ESTADÍSTICAS DE SUMA POR TIPO DE POKEMON

```python
suma_tipo = df_numeric.groupby(df["Type 1"]).sum()
print(suma_tipo)
```

(Agrupación - groupBy) ESTADÍSTICAS DE NÚMERO DE POKEMONS POR TIPO 1DE POKEMON

```python
count_tipo = df.groupby(df["Type 1"]).count()
print(count_tipo)
```

(Agrupación - groupBy) ESTADÍSTICAS DE NÚMERO DE POKEMONS POR TIPO 1 y 2 DE POKEMON

```python
count_tipo_2 = df.groupby(['Type 1', 'Type 2']).size().reset_index(name='Count')
print(count_tipo_2)
```

Nota: SI TENEMOS ARCHIVOS ENORMES (1TB) PODEMOS LEERLOS POR PARTES Cada fila podría estar acumulando cerca de 20 bytes, por lo que podríamos estar trabajando con cantidades enormes

LEE EL ARCHIVO CVS SEPARÁNDOLO POR CHUNKS Y CON UN TAMAÑO DE (chunksize=5)

ITERA POR LOS CHUNKS Y MUÉSTRALOS POR CONSOLA

```python
for chunk in pd.read_csv('pokemon_data.csv', chunksize=5):
    print(chunk)
```
